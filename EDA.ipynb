```python
import altair as alt
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(12, 6))
sns.histplot(precio_surtidor['precio'], bins=30, kde=True)
plt.title('Distribución de Precios - Precio Surtidor')
plt.xlabel('Precio')
plt.ylabel('Frecuencia')
```

```python
plt.figure(figsize=(16, 6))
sns.boxplot(x='provincia', y='precio', data=precio_surtidor)
plt.title('Boxplot de Precios de Surtidores de Combustible')
plt.xlabel('Provincia')
plt.ylabel('Precio')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
sns.boxplot(y='precio', data=precio_surtidor)
plt.title('Boxplot General de Precios de Surtidores de Combustible')
plt.ylabel('Precio')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.violinplot(y='precio', data=precio_surtidor)
plt.title('Violin Plot de Precios - Precio Surtidor')
plt.ylabel('Precio')
plt.show()
```
