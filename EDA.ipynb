```python
plt.figure(figsize=(12, 6))
sns.histplot(precio_surtidor['precio'], bins=30, kde=True)
plt.title('Distribucin de Precios - Precio Surtidor')
plt.xlabel('Precio')
plt.ylabel('Frecuencia')
```

```python
precio_surtidor.groupby('producto')['precio'].agg(['mean', 'median']).sort_values(by='mean', ascending=False).head(10)
```

```python
def detectar_outliers(df):
    Q1 = df['precio'].quantile(0.20) #en menos de 0.2 no hay
    Q3 = df['precio'].quantile(0.80) #en mas de 0.8 no hay
    IQR = Q3 - Q1
    return df[(df['precio'] < (Q1 - 1.5 * IQR)) | (df['precio'] > (Q3 + 1.5 * IQR))]

outliers_surtidor = detectar_outliers(precio_surtidor)
print(outliers_surtidor)
```

```python
plt.figure(figsize=(12, 6))
plt.scatter(precio_surtidor['fecha_vigencia'], precio_surtidor['precio'], label='Precio Surtidor', alpha=0.5)
plt.scatter(outliers_surtidor['fecha_vigencia'], outliers_surtidor['precio'], color='red', label='Outliers', marker='o')
plt.title('Precio Surtidor y Valores Atpicos')
plt.xlabel('Fecha')
plt.ylabel('Precio')
plt.legend()
plt.show()
```
