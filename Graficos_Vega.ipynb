```python
import pandas as pd
import folium
from folium.plugins import HeatMap
```

```python
# Create a Folium map centered around the average location
latitude_center = precio_historico['latitud'].mean()
longitude_center = precio_historico['longitud'].mean()
m = folium.Map(location=[latitude_center, longitude_center], zoom_start=10)

# Prepare data for the heatmap
# We need a list of [latitude, longitude, value] for each point
heat_data = [[row['latitud'], row['longitud'], row['precio']] for index, row in precio_historico.iterrows()]

# Add the heatmap layer
HeatMap(heat_data).add_to(m)

# Display the map
m.save('heatmap.html')
m
```
